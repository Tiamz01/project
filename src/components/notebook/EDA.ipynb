 ## brief explanation of the columns in your dataset:

1. **CountryID:** An identifier for each country.

2. **Country:** The name of the country being represented in the data.

3. **Year:** The year for which the economic data is recorded.

4. **AMA Exchange Rate:** Exchange rate data based on the AMA (Averaging Moving Average) method.

5. **IMF-Based Exchange Rate:** Exchange rate data based on the International Monetary Fund's method.

6. **Population:** The total population of the country in the given year.

7. **Currency:** The official currency of the country.

8. **Per Capita GNI:** The Gross National Income per capita, which is the total GNI of the country divided by its population.

9. **Agriculture, Hunting, Forestry, Fishing (ISIC A-B):** Value-added by economic activities in the agriculture sector.

10. **Changes in Inventories:** The change in the value of inventories held by businesses in the country.

11. **Construction (ISIC F):** Value-added by the construction sector.

12. **Exports of Goods and Services:** The total value of goods and services exported by the country.

13. **Final Consumption Expenditure:** Total expenditure on consumption by households, government, and non-profit institutions.

14. **General Government Final Consumption Expenditure:** Expenditure by the government on consumption.

15. **Gross Capital Formation:** The total value of investments in the country, including changes in inventories.

16. **Gross Fixed Capital Formation (including Acquisitions less Disposals of Valuables):** Investment in fixed assets, excluding disposals of valuables.

17. **Household Consumption Expenditure (including Non-profit Institutions Serving Households):** Total expenditure by households and non-profit institutions.

18. **Imports of Goods and Services:** The total value of goods and services imported by the country.

19. **Manufacturing (ISIC D):** Value-added by economic activities in the manufacturing sector.

20. **Mining, Manufacturing, Utilities (ISIC C-E):** Value-added by economic activities in the mining, manufacturing, and utilities sectors.

21. **Other Activities (ISIC J-P):** Value-added by various other economic activities in sectors with ISIC codes ranging from J to P.

22. **Total Value Added:** The total value added by all economic activities in the country.

23. **Transport, Storage, and Communication (ISIC I):** Value-added by economic activities in the transport, storage, and communication sector.

24. **Wholesale, Retail Trade, Restaurants, and Hotels (ISIC G-H):** Value-added by economic activities in the wholesale, retail trade, restaurants, and hotels sector.

25. **Gross National Income (GNI) in USD:** The total GNI of the country in US dollars.

26. **Gross Domestic Product (GDP):** The total GDP of the country, representing the overall economic output.

```python
cols_to_drop = ['CountryID', 'Currency', 'AMA_exchange_rate', 'Changes_in_inventories']

fixed_data.drop(columns=cols_to_drop, inplace=True, axis=1)
fixed_data
```

```python
col_2_rename = {'IMF_based_exchange_rate':'exchange_rate'}
fixed_data = fixed_data.rename(columns=col_2_rename)
fixed_data.columns
```

## Fixing missing value

```python
fixed_data.isna().sum()
```

```python
fixed_data['Agriculture_hunting_forestry_fishing'].fillna(np.mean(fixed_data['Agriculture_hunting_forestry_fishing']), inplace=True)
```

```python
fixed_data['Exports_of_goods_and_services'].fillna(np.mean(fixed_data['Exports_of_goods_and_services']), inplace=True)

fixed_data['General_government_final_consumption_expenditure'].fillna(np.mean(fixed_data['General_government_final_consumption_expenditure']), inplace=True)
fixed_data['Gross_capital_formation'].fillna(np.mean(fixed_data['Gross_capital_formation']), inplace=True)
fixed_data['Gross_fixed_capital_formation'].fillna(np.mean(fixed_data['Gross_fixed_capital_formation']), inplace=True)
fixed_data['Household_consumption_expenditure'].fillna(np.mean(fixed_data['Household_consumption_expenditure']), inplace=True)
fixed_data['Imports_of_goods_and_services'].fillna(np.mean(fixed_data['Imports_of_goods_and_services']), inplace=True)
fixed_data['Manufacturing_index'].fillna(np.mean(fixed_data['Manufacturing_index']), inplace=True)
fixed_data['Transport_storage_and_communication'].fillna(np.mean(fixed_data['Transport_storage_and_communication']), inplace=True)
fixed_data['Wholesale_retail_trade_restaurants_and_hotels'].fillna(np.mean(fixed_data['Wholesale_retail_trade_restaurants_and_hotels']), inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
GDP_mean = data_to_use['Gross_Domestic_Product'].mean()
sorted_GDP = data_to_use.sort_values(by='Gross_Domestic_Product', ascending=False)
sorted_GDP_20 = sorted_GDP.head(20)
sorted_GDP_20.columns
```

## The top countries that has been doing good on gdp as seen to be china and usa. These countries lead in country with highest occuring time of gdp values

```python
# Create a horizontal bar plot for the top 20 countries with the highest GDP
plt.barh(sorted_GDP_20['Country'], sorted_GDP_20['Gross_Domestic_Product'])
plt.xlabel('Gross Domestic Product')
plt.ylabel('Country')
plt.title('Most occurring Countries with the Highest GDP')
plt.show()

```

```python
sorted_GDP_last_20 = sorted_GDP.tail(20)
sorted_GDP_last_20
```

```python
# Create a horizontal bar plot for the top 20 countries with the highest GDP

plt.barh(sorted_GDP_20['Country'], sorted_GDP_last_20['Gross_Domestic_Product'])
plt.xlabel('Gross Domestic Product')
plt.ylabel('Country')
plt.title('Most occuring Countries with the least GDP')
plt.show()
```

```python
# Create an empty figure and axis
fig, ax = plt.subplots(figsize=(15,12))

# Grouping the data by 'Year' and select the top 20 countries with the highest GDP for each year
GDP_by_year = sorted_GDP.groupby('Year').apply(lambda x: x.nlargest(20, 'Gross_Domestic_Product')).reset_index(drop=True)

# Create a horizontal bar plot for each year
for year, group in GDP_by_year.groupby('Year'):
    ax.barh(group['Country'], group['Gross_Domestic_Product'], label=year)

# Set labels and title
ax.set_xlabel('Gross Domestic Product')
ax.set_ylabel('Country')
ax.set_title('Top 20 Countries with the Highest GDP by Year')

# Show a legend to distinguish different years
ax.legend(title='Year', loc='upper right', bbox_to_anchor=(1.1, 1.0))

plt.tight_layout()
plt.show()
```
